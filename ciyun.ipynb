```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
d = "/home/qxliang/word_cloud-master/examples"  #本地系统文件操作需要的路径
```

```python
text = open(path.join(d, 'keywords.txt')).read()  #读取整个文本
```

```python
text
```

```python
alice_mask = np.array(Image.open(path.join(d, "alice_color.png"))) #将图片转化为numpy数组
```

```python
alice_mask  
```

```python
wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,width=3) #更多参数参考WordCloud模块的帮助说明
```

```python
wc.generate(text)  #生成词云对象
```

```python
wc.to_file(path.join(d, "alice2.png")) #保存到想要的文件
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#显示可交互的图像
%matplotlib notebook
from ipywidgets import *
def f():
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
interact(f)
```
